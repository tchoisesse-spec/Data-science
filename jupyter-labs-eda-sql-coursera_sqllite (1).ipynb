## Tasks

Now write and execute SQL queries to solve the assignment tasks.

**Note: If the column names are in mixed case enclose it in double quotes
   For Example "Landing_Outcome"**

### Task 1




##### Display the names of the unique launch sites  in the space mission


```python
# Display the names of the unique launch sites
query_1 = "SELECT DISTINCT Launch_Site FROM SPACEXTABL;"
pd.read_sql_query(query_1, con)
```


### Task 2


#####  Display 5 records where launch sites begin with the string 'CCA' 


```python
# Display 5 records where launch sites begin with the string 'CCA'
query_2 = "SELECT * FROM SPACEXTABL WHERE Launch_Site LIKE 'CCA%' LIMIT 5;"
pd.read_sql_query(query_2, con)
```

### Task 3




##### Display the total payload mass carried by boosters launched by NASA (CRS)


```python
# Display the total payload mass carried by boosters launched by NASA (CRS)
query_3 = "SELECT SUM(Payload_Mass__kg_) AS Total_Payload_Mass FROM SPACEXTABL WHERE Customer = 'NASA (CRS)';"
pd.read_sql_query(query_3, con)
```

### Task 4




##### Display average payload mass carried by booster version F9 v1.1


```python
# Display average payload mass carried by booster version Falcon 9
query_4 = "SELECT AVG(Payload_Mass__kg_) AS Average_Payload_Mass FROM SPACEXTABL WHERE Booster_Version LIKE 'Falcon 9%';"
pd.read_sql_query(query_4, con)
```

### Task 5

##### List the date when the first succesful landing outcome in ground pad was acheived.


_Hint:Use min function_ 


```python
# List the date when the first successful landing outcome in ground pad was achieved
query_5 = "SELECT MIN(Date) AS First_Success_Ground_Pad FROM SPACEXTABL WHERE Landing_Outcome = 'Success (ground pad)';"
pd.read_sql_query(query_5, con)
```

### Task 6

##### List the names of the boosters which have success in drone ship and have payload mass greater than 4000 but less than 6000


```python
# List the names of the boosters which have success in drone ship and have payload mass greater than 4000 but less than 6000
query_6 = """
SELECT Booster_Version FROM SPACEXTABL 
WHERE Landing_Outcome = 'Success (drone ship)' 
AND Payload_Mass__kg_ > 4000 
AND Payload_Mass__kg_ < 6000;
"""
pd.read_sql_query(query_6, con)
```

### Task 7




##### List the total number of successful and failure mission outcomes


```python
# List the total number of successful and failure mission outcomes
query_7 = "SELECT Mission_Outcome, COUNT(*) AS Total_Count FROM SPACEXTABL GROUP BY Mission_Outcome;"
pd.read_sql_query(query_7, con)
```

### Task 8



##### List all the booster_versions that have carried the maximum payload mass, using a subquery with a suitable aggregate function.


```python
# Task 8: List the names of the booster_versions which have carried the maximum payload mass
query_8 = "SELECT Booster_Version FROM SPACEXTABL WHERE Payload_Mass__kg_ = (SELECT MAX(Payload_Mass__kg_) FROM SPACEXTABL);"
pd.read_sql_query(query_8, con)
```

### Task 9


##### List the records which will display the month names, failure landing_outcomes in drone ship ,booster versions, launch_site for the months in year 2015.

**Note: SQLLite does not support monthnames. So you need to use  substr(Date, 6,2) as month to get the months and substr(Date,0,5)='2015' for year.**


```python
# List the failed landing_outcomes in drone ship, their booster versions, and launch site names for the year 2015
query_9 = """
SELECT Date, Booster_Version, Launch_Site, Landing_Outcome FROM SPACEXTABL 
WHERE Landing_Outcome = 'Failure (drone ship)' 
AND Date LIKE '2015%';
"""
pd.read_sql_query(query_9, con)
```

### Task 10




##### Rank the count of landing outcomes (such as Failure (drone ship) or Success (ground pad)) between the date 2010-06-04 and 2017-03-20, in descending order.


```python
#  Rank the count of landing outcomes between the date 2010-06-04 and 2017-03-20, in descending order
query_10 = """
SELECT Landing_Outcome, COUNT(*) AS Outcome_Count FROM SPACEXTABL 
WHERE Date BETWEEN '2010-06-04' AND '2017-03-20' 
GROUP BY Landing_Outcome 
ORDER BY Outcome_Count DESC;
"""
pd.read_sql_query(query_10, con)
```
